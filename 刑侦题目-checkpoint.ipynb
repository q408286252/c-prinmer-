```python
def p2ti(l):
    if l[5] == "c":
        l[2] = "a"
        #print("第二题:%s" %l[2])
    elif l[5] == "d":
        l[2] = "b"
        #print("第二题:%s" %l[2])
    elif l[5] == "a":
        l[2] = "c"
        #print("第二题:%s" %l[2])
    elif l[5] == "b":
        l[2] = "d"
        #print("第二题:%s" %l[2])
    else:
        l[2] = 0
    return l

def p3ti (l):
    if l[4] != 0 and l[6] != 0 and l[2] != 0 :
        l[3] = "a"
        if l[4] == l[6] == l[2] != l[3]:
            #print("第三题:%s" %l[3])
            return l
        l[3] = "b"
        elif l[4] == l[3] == l[2] != l[6]:
            #print("第三题:%s" %l[3])
            return l
        l[3] = "c"
        elif l[4] == l[6] == l[3] != l[2]:
            #print("第三题:%s" %l[3])
            return l
        l[3] = "d"
        elif l[3] == l[6] == l[2] != l[4]:
            #print("第三题:%s" %l[3])
            return l
        else:
            l[3] = 0
    return l

def p4ti (l): #有关7
    if l[1] == l[5] and l[2] != l[7] and l[1] != l[9] and l[6] != l[10]:
        l[4] = "a"
        #print("第四题:%s" %l[4])
    elif l[1] != l[5] and l[2] == l[7] and l[1] != l[9] and l[6] != l[10]:
        l[4] = "b"
       # print("第四题:%s" %l[4])
    elif l[1] != l[5] and l[2] != l[7] and l[1] == l[9] and l[6] != l[10]:
        l[4] = "c"
        #print("第四题:%s" %l[4])
    elif l[1] != l[5] and l[2] != l[7] and l[1] != l[9] and l[6] == l[10]:
        l[4] = "d"
        #print("第四题:%s" %l[4])
    else:
        l[4] = 0
    return l

def p5ti (l):
    if l[8] == "a" and l[4] != "a" and l[9] != "a" and l[7] != "a":
        l[5] = "a"
        #print("第五题:%s" %l[5])
    elif l[4] == "b" and l[8] != "b" and l[9] != "b" and l[7] != "b":
        l[5] = "b"
        #print("第五题:%s" %l[5])
    elif l[9] == "c" and l[4] != "c" and l[8] != "c" and l[7] != "c":
        l[5] = "c"
        #print("第五题:%s" %l[5])
    elif l[7] == "d" and l[9] != "d" and l[4] != "c" and l[8] != "d":
        l[5] = "d"
        #print("第五题:%s" %l[5])
    else:
        l[5] = 0
    return l
    
def p6ti (l):
    if l[8] != 0:
        if l[2] == l[4] == l[8]:
            if l[1] != l[8] or l[6] != l[8]:
                if l[3] != l[8] or l[10] != l[8]:
                    if l[5] != l[8] or l[9] != l[8]:
                        l[6] = "a"
            #print("第六题:%s" %l[6])
        if l[2] != l[4] != l[8]:
            if l[1] == l[8] or l[6] == l[8]:
                if l[3] != l[8] or l[10] != l[8]:
                    if l[5] != l[8] or l[9] != l[8]:
                        l[6] = 'b'
            #print("第六题:%s" %l[6])
        if l[2] != l[4] != l[8]:
            if l[1] != l[8] or l[6] != l[8]:
                if l[3] == l[8] or l[10] == l[8]:
                    if l[5] != l[8] or l[9] != l[8]:
                        l[6] = 'c'
            #print("第六题:%s" %l[6])
        if l[2] != l[4] != l[8]:
            if l[1] != l[8] or l[6] != l[8]:
                if l[3] != l[8] or l[10] != l[8]:
                    if l[5] == l[8] or l[9] == l[8]:
                        l[6] = 'd'
            #print("第六题:%s" %l[6])
    return l

def p7ti(l):
    a = 0; b = 0; c = 0; d = 0
    for i in l:
        if i == 'a':
            a += 1
        elif i == "b":
            b += 1
        elif i == "c":
            c += 1
        elif i == "d":
            d += 1
    if c == min(a, b, c, d) and c >= 1 and c != a and c != b and c != d:
        l[7] = "a"
        #print("第七题:%s" %l[7])
    elif b == min(a, b, c, d) and b >= 1 and b != a and b != c and b != d:
        l[7] = "b"
        #print("第七题:%s" %l[7])
    elif a == min(a, b, c, d) and a >= 1 and a != b and a != c and a != d:
        l[7] = "c"
        #print("第七题:%s" %l[7])
    elif d == min(a, b, c, d) and d >= 1 and d != a and d != b and d != c:
        l[7] = "d"
        #print("第七题:%s" %l[7])
    else:
        l[7] = 0
    return l  
    
def p8ti(l): #有关7
    if l[1] != 0:
        if l[1] == "a":
            left = right = "b"
        elif l[1] == "b":
            left = "a"; right = "c"
        elif l[1] == "c":
            left = "b"; right = "d"
        elif l[1] == "d":
            left = right = "c"
    if l[7] != 0 and l[7] != left and l[7] != right:
        l[8] = "a"
        #print("第八题:%s" %l[8])
    if l[5] != 0 and l[5] != left and l[5] != right:
        l[8] = "b"
        #print("第八题:%s" %l[8])
    if l[2] != 0 and l[2] != left and l[2] != right:
        l[8] = "c"
        #print("第八题:%s" %l[8])
    if l[10] != 0 and l[10] != left and l[10] != right:
        l[8] = "d"
        #print("第八题:%s" %l[8])
    return l

def p9ti(l):
    if l[1] != 0 and l[6] != 0 and l[5] != 0:
        if l[1] == l[6]:
            if l[6] != 0 and l[6] != l[5]:
                l[9] = "a"
                #print("第九题:%s" %l[9])
            if l[10] != 0 and l[10] != l[5]:
                l[9] = "b"
                #print("第九题:%s" %l[9])
            if l[2] != 0 and l[2] != l[5]:
                l[9] = "a"
                #print("第九题:%s" %l[9])
            if l[5] != "d":   #第九题与第五题答案不相同
                l[9] = "d"
                #print("第九题:%s" %l[9])
        else:
            if l[6] != 0 and l[6] == l[5]:
                l[9] = "a"
                #print("第九题:%s" %l[9])
            if l[10] != 0 and l[10] == l[5]:
                l[9] = "b"
                #print("第九题:%s" %l[9])
            if l[2] != 0 and l[2] == l[5]:
                l[9] = "a"
                #print("第九题:%s" %l[9])
            if l[5] == "d":   #第九题与第五题答案相同
                l[9] = "d"
                #print("第九题:%s" %l[9])
    return l

def p10ti(l):
    a = 0; b = 0; c = 0; d = 0
    for i in l:
        if i == 'a':
            a += 1
        elif i == "b":
            b += 1
        elif i == "c":
            c += 1
        elif i == "d":
            d += 1
    ca = max(a, b, c, d) - min(a, b, c, d)
    if ca == 3:
        l[10] = "a"
        #print("第十题:%s" %l[10])
    if ca == 2:
        l[10] = "b"
        #print("第十题:%s" %l[10])
    if ca == 4:
        l[10] = "c"
        #print("第十题:%s" %l[10])
    if ca == 1:
        l[10] = "d"
        #print("第十题:%s" %l[10])
    return l
import copy
la = []
a =[[0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10] for a1 in ["a","b","c","d"] for a2 in ["a","b","c","d"] for a3 in ["a","b","c","d"] for a4 in ["a","b","c","d"] for a5 in ["a","b","c","d"] for a6 in ["a","b","c","d"] for a7 in ["a","b","c","d"] for a8 in ["a","b","c","d"] for a9 in ["a","b","c","d"] for a10 in ["a","b","c","d"]]
for i in a:
    i_new = copy.copy(i)
    p2ti(i)
    p3ti(i)
    p4ti(i)
    p5ti(i)
    p6ti(i)
    p8ti(i)
    p9ti(i)
    p7ti(i)
    p10ti(i)
    if i == i_new:
        la.append(i)
        print(la)
```
